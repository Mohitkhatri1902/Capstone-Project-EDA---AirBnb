#### Chart - 8

**Q8) The name column has given which property name is mentioned so that we can find how many times each apartment has been booked, considering room type, so that we analyzed which is more revenue-generating and which is less?**

````python
# Chart - 8 visualization code

# Calculate booking count for each property and room type
booking_count = AirBnb_df.groupby(['name','room_type']).size().reset_index(name='booking_count')

# Calculate total revenue for each property based on 'price' and 'booking_count'
AirBnb_df['total_revenue'] = AirBnb_df['price'] * booking_count['booking_count']
property_revenue = AirBnb_df.groupby(['name', 'room_type'])['total_revenue'].sum().reset_index()

# Sort by booking count and total revenue to identify top and bottom performers
sort_by_booking_count = booking_count.sort_values(by='booking_count', ascending=False)
sort_by_total_revenue = property_revenue.sort_values(by='total_revenue', ascending=False)
print("Top Performance by Bokking Count:")
print(sort_by_booking_count.head(10))
print("\nTop Performance By Total Revenue:")
print(sort_by_total_revenue.head(10))
````

````python
print("Top Performance by Bokking Count:")
sort_by_booking_count.head(10)
````

````python
# Create a grouped bar chart for booking count
plt.figure(figsize=(10,6))
top_10 = sort_by_booking_count.head(10)
plt.barh(top_10['name'], top_10['booking_count'], color='skyblue')
plt.xlabel('Booking Count')
plt.ylabel('Property Name')
plt.title('Top Properties by Booking Count')
plt.gca().invert_yaxis()
plt.xticks(rotation=90)
plt.show()
````

````python
# Create a Horizontal bar chart for Total Revenue
plt.figure(figsize=(10,6))
top_10 = sort_by_total_revenue.head(10)
plt.barh(top_10['name'], top_10['total_revenue'], color='lightcoral')
plt.xlabel('Total Revenue')
plt.ylabel('Property Name')
plt.title('Top Properties by Total Revenue')
plt.gca().invert_yaxis()
plt.xticks(rotation=90)
plt.show()
````

#### Chart - 10

#Q10) Find total Numbers of nights spend as per location

````python
# Chart - 10 visualization code

#find the maximum(minimum nights) spending by customer across different neighbourhood group
total_nights = airbnb.groupby('neighbourhood_group')['minimum_nights'].sum().reset_index()
final_nights = total_nights.sort_values('minimum_nights',ascending=True)
final_nights.head()
````

````python
#visualization of total number of nights spend per location using seaborn
sns.set(rc={'figure.figsize':(10,5)})
sns.set_style('white')
order = final_nights.groupby('neighbourhood_group')['minimum_nights'].mean().sort_values(ascending=False).index
nights = sns.barplot(x='neighbourhood_group',y='minimum_nights',data=final_nights)
nights.set_title('Total No. of Nights Spend per Location', weight='bold')
nights.set_xlabel('neighbourhood_group',weight='bold')
nights.set_ylabel('minimum_nights', weight ='bold')
nights.set_xticklabels(nights.get_xticklabels(),weight='bold',rotation=45, size='15')
````

````python
# Define the desired order for 'neighbourhood_group'
desired_order = final_nights.groupby('neighbourhood_group')['minimum_nights'].mean().sort_values(ascending=False).index

# Convert 'neighbourhood_group' to a categorical data type with the desired order
final_nights['neighbourhood_group'] = pd.Categorical(final_nights['neighbourhood_group'], categories=desired_order, ordered=True)

sns.set(rc={'figure.figsize':(10,5)})
sns.set_style('white')

# Create the bar plot
nights = sns.barplot(x='neighbourhood_group', y='minimum_nights', data=final_nights)
nights.set_title('Total No. of Nights Spend per Location', weight='bold')
nights.set_xlabel('neighbourhood_group', weight='bold')
nights.set_ylabel('minimum_nights', weight='bold')
nights.set_xticklabels(nights.get_xticklabels(), weight='bold', rotation=45, size='15')

plt.show()
````
